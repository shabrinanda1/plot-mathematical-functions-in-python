# Plotting Mathematical functions

```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

## 1.  Plotting the function 'y=x^2'

```python
def plot_xsquare():  
    x=np.linspace(-10, 10, 21)
    #print(x)
    y=[]
    for i in x:
        c=i*i
        y.append(c)
        #print(y)
    plt.plot(x,y)
    plt.xlabel('X AXIS')
    plt.ylabel('Y AXIS')
    plt.title('PLOT OF THE FUNCTION Y=X^2')
    plt.axhline(y=0, color='green', linewidth=0.9)
    plt.axvline(x=0, color='green', linewidth=0.9)
    plt.grid()
    plt.show()
```

```python
plot_xsquare()
```

## 2.  Plotting of the function 'y=x^3'

```python
def plot_xcube():
    x=np.linspace(-10, 10, 21)
    y=[]
    for i in x: 
        c=i*i*i
        y.append(c)   
    #print(y)   
    plt.plot(x,y)
    plt.xlabel('X AXIS')
    plt.ylabel('Y AXIS')
    plt.title('PLOT OF THE FUNCTION Y=X^3')
    plt.axhline(y=0, color='green', linewidth=0.9)
    plt.axvline(x=0, color='green', linewidth=0.9)
    plt.grid()
    plt.show()
```

```python
plot_xcube()
```

## 3.   Plotting of  the function 'y=1/x'

```python
def plot_reciprocal():
    x1=np.linspace(-10, -1, 20)
    x2=np.linspace(1, 10, 20)
    y1=[]
    for i in x1:
        c=1/i
        y1.append(c)  
    # print(y1) 
    y2=[]
    for i in x2:
        c=1/i
        y2.append(c)   
    #print(y2)   
    plt.plot(x2,y2)
    plt.plot(x1,y1)
    plt.xlabel('X AXIS')
    plt.ylabel('Y AXIS')
    plt.title('PLOT OF THE FUNCTION Y=1/X')
    plt.axhline(y=0, color='green', linewidth=0.9)
    plt.axvline(x=0, color='green', linewidth=0.9)
    plt.grid()
    plt.show()
```

```python
plot_reciprocal()
```

## 4.   Plotting of the function 'y=mod x'

```python
def plot_modX():
    x=np.linspace(-10, 10, 21)
    #print(x)
    y=[]
    for i in x: 
        c=abs(i)
        y.append(c) 
    # print(y)
    plt.plot(x,y)
    plt.xlabel('X AXIS')
    plt.ylabel('Y AXIS')
    plt.title('PLOT OF THE FUNCTION Y= ABSOLUTE VALUE OF X')
    plt.axhline(y=0, color='green', linewidth=0.9)
    plt.axvline(x=0, color='green', linewidth=0.9)
    plt.grid()
    plt.show()
```

```python
plot_modX()
```

## 5.   Plotting of the function 'y=e^x'


```python
def plot_exponent():
    x=np.linspace(-10, 10, 200)
    y = np.exp(x)
    # print(y)
    plt.plot(x,y)
    plt.xlabel('X AXIS')
    plt.ylabel('Y AXIS')
    plt.title('PLOT OF THE FUNCTION Y=e^x')
    plt.axhline(y=0, color='green', linewidth=0.9)
    plt.axvline(x=0, color='green', linewidth=0.9)
    plt.grid()
    plt.show()
```

```python
plot_exponent()
```
